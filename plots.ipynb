```python
import sys
import os
import matplotlib.pyplot as plt


def read_file(file_path):
    with open(file_path, "r") as file:
        values = []
        lines = file.readlines()

        for line in lines:
            value = float(line.split()[-1])
            values.append(value)

    return values


```

```python
file_paths = os.listdir()
x = list(range(1, 26))
for path in file_paths:
    if ".loss" in path:
        if "-40." in path or "-60." in path:
            data = read_file(path)
            plt.scatter(x, data, label=f"{path}", c="red")

        if "-20." in path:
            data = read_file(path)
            plt.scatter(x, data, label=f"{path}", c="red")
# plt.ylim(0, 10)
plt.xlabel("Iteration number")
plt.xlabel("Loss")
plt.legend()
plt.show()
```
